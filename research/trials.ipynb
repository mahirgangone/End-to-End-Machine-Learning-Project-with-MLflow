```python
def get_product(x:int, y:int) -> int:
    return x * y

get_product(2, 3)
```

```python
get_product(2, "4")
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product(x:int, y:int) -> int:
    return x * y

get_product(2, 4)
```

```python
#Ensures value type is an integer
get_product(2, "4")
```
